```python
def find_test_score(row):
    key = df.loc[0, "Q1":"Q27"]
    points = 0
    count = 0
    #TODO: Compare test answers with key and get test score
    for numCols in key:
        count = count + 1
        currentCol = 'Q' + str(count)
        value1 = key[currentCol]
        value2 = row[currentCol]
        if value1 == value2:
            points = points + 1   
    return (points/27)*100

count = 0
rowNum = 2
while count < 147:
    count = count + 1
    tempRow = df.loc[rowNum, "Q1":"Q27"]
    score = find_test_score(tempRow)
    df.loc[rowNum, 'TestScore'] = score
    rowNum = rowNum + 1
    
    
df['TestScore'].astype(int)
df.round(2)
```

```python
def sort_grades(column):
    A = 0
    B = 0
    C = 0
    D = 0
    F = 0
    for value in column:
        if value >= 90 and value <= 100:
            A = A + 1
        elif value >= 80 and value <= 89.9:
            B = B + 1
        elif value >= 70 and value <= 79.9:
            C = C + 1
        elif value >= 60 and value <= 69.9:
            D = D + 1
        else:
            F = F + 1
    print("   Number of students with a 90-100 grade: " + str(A))
    print("   Number of students with a 80-89 grade: " + str(B))
    print("   Number of students with a 70-79 grade: " + str(C))
    print("   Number of students with a 60-69 grade: " + str(D))
    print("   Number of students with a grade of 59 or lower: " + str(F))
```

```python
# TEST SCORE VS FINAL COURSE GRADE:
```

```python
AvgTestScore = df.loc[2 : 148, 'TestScore'].mean().round(2)
TestScoresOverall = df.loc[2 : 148, 'TestScore']

AvgPercentGrade = df['PercentGrade'].mean().round(2)
FinalGradesOverall = df.loc[2 : 148, 'PercentGrade']

print("Average Test Score: " + str(AvgTestScore))
print("Test Scores Overall: ")
sort_grades(TestScoresOverall)

print("\n")

print("Average Final Course Grade: " + str(AvgPercentGrade))
print("Final Course Grades Overall: ")
sort_grades(FinalGradesOverall)
```

```python
# Conclusion:

# A student's test score and final course grade are unrelated. All except one of the test scores are in the F range,
# while final course grades are more evenly spread out across the board. Therefore, a poor test grade does not mean
# a student will get a poor final course grade.
```
